## 1. SALES: UNITS, VOLUME, VALUE

```python
#Plot Aggregate Sales and Volume
sns.set(style="white")
f, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(12, 12), sharex=True)

# Plot Unit Sales
sns.lineplot(x = "DATE", y = "Vtas (Uds)", data = fact_data, ax=ax1)
ax1.set_ylabel("Sales: Units")

# Plot Volume Sales
sns.lineplot(x = "DATE", y = "Vtas (Vol)", data = fact_data, ax=ax2)
ax2.set_ylabel("Sales: Volume")

# Plot Value Sales
sns.lineplot(x = "DATE", y = "Vtas (Valor)", data = fact_data, ax=ax3)
ax3.set_ylabel("Sales: Value")
```

Cyclical peaks during the holidays, less prominent in the holiday season 2019. Highest peak COVID March-April 2020. Will use Sales Value on rest of the analysis for now since it is the one with highest variance. 


## 2. MKT TYPE

```python
mkt_type = fact_data[fact_data['MKT_TAG'].isin(['ORG H+S', 'SUPER', 'TOTAL ONLINE', 'EESS', 'TOTAL SUPER + HIPER P&B', 'HIPER', 'PERFUMERIAS', 'EES+INDEP', 'ESPANA+EESS'])]
```

```python
plt.figure(figsize=(15,8))
g = sns.relplot(data=mkt_type,
                x="DATE", 
                y="Vtas (Valor)", 
                hue='MKT_TAG', 
                kind='line', 
                err_style=None, 
                height=10, 
                aspect=3)
g.fig.suptitle('Value of Sales by Market Type')
```

All categories follow a very close pattern. Top: Total Online, Super>Hiper, Bottom: Perfumerias seems to develop 2-3 peaks during holiday season, Hiper, EES Independiente.  Would be interesting to explore **Online**.

```python
plt.figure(figsize=(15,8))
g = sns.countplot(data=mkt_type,
                x="MKT_TAG",  
                hue='YEAR',
                height=10, 
                aspect=3)
g.set_title('Market Type by Year')
```

Interesting that most categories carry same count of transactions, maybe because of how the data was split? Or fix number of companies reporting to Nielsen. ORG H+S has the highest transaction count. 2020 is not a full year yet. 

```python
plt.figure(figsize=(15,8))
g = sns.catplot(data=mkt_type,
                x="MKT_TAG", 
                y="Vtas (Valor)", 
                hue='YEAR', 
                kind='bar',  
                height=10, 
                aspect=3)
g.fig.suptitle('Value of Sales by Market Type')
```

Total volume of sales year to year seems stable meaning sales remain fairly constant in most categories when looking at the entire year. Check why total online is at the bottom? Check code later. 

## 3. MKT POP

```python
mkt_pop = fact_data[fact_data['MKT_TAG'].isin(['50.000 - 100.000 hab.', '10.000 - 20.000 hab.', '>500.000 hab.', 'Indep. <300', '< 10.000 hab.', '20.000 - 50.000 hab.', '100.000 - 500.000 hab.'])]
```

```python
plt.figure(figsize=(15,8))
g = sns.relplot(data=mkt_pop,
                x="DATE", 
                y="Vtas (Valor)", 
                hue='MKT_TAG', 
                kind='line', 
                err_style=None, 
                height=10, 
                aspect=3)
g.fig.suptitle('Value of Sales by Market Population')
```

'Indep <300' does not see fluctuations like the other categories. Outliers: 20-50 and <10 wider curve mid year, >500 seems to have opposite pattern and decline midway through the year when everyone else is either stable or growing. 

```python
plt.figure(figsize=(15,8))
g = sns.countplot(data=mkt_pop,
                x="MKT_TAG",  
                hue='YEAR')
g.set_title('Market Population by Year')
```
